2nd order Polynomial regression

```python
#Adding bias as one
x1=np.array(xtrain).reshape(100,1)
y1=np.array(ytrain).reshape(100,1)
one=np.ones(100)
one=np.array(one).reshape(100,1)

x1=np.append(one,x1,axis=1)
x2=np.append(one,x2,axis=1)
```

```python
#Taking square of the train and test data
x_square=np.power(xtrain,2)
x2_square= np.power(xtest,2)         
x_square=np.array(x_square).reshape(100,1)
x2_square= np.array(x2_square).reshape(100,1) 
X=np.append(x1,x_square,axis=1)
X2=np.append(x2,x2_square,axis=1)
#normal equation to calculate theta (in one line)
theta2 = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y1)
y_pred2 = X.dot(theta2)
#y_predt2 = X2.dot(theta2)
# X = np.array([X, np.power(X,2), X**3])
plt.scatter(xtrain,y_pred2)
plt.scatter(xtrain,ytrain)
```

3rd order Polynomial regression

```python
#Taking 3rd order of the train and test data
x_cube=np.power(xtrain,3)
x3_cube= np.power(xtest,3)         
x_cube=np.array(x_cube).reshape(100,1)
x3_cube= np.array(x3_cube).reshape(100,1) 
X=np.append(x1,x_cube,axis=1)
X2=np.append(x2,x3_cube,axis=1)
#normal equation to calculate theta (in one line)
theta3 = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y1)
y_pred3 = X.dot(theta3)
#y_predt2 = X2.dot(theta2)
# X = np.array([X, np.power(X,2), X**3])
plt.scatter(xtrain,y_pred3)
plt.scatter(xtrain,ytrain)
```
